```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[0,2,4,6]
plt.plot(x,y)
plt.title('Our First Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xticks([1,2,3,4])
plt.yticks([0,2,4,6])
plt.legend(['Our'])# if there are many lines it helps in distinguishing between them.
plt.show()
```

```python
x=[1,2,3,4]
y=[0,2,4,6]
print(plt.style.available)
plt.style.use('seaborn-deep')
plt.plot(x,y,'y--',linewidth=3,label='Our') #FormatString=[marker][linestyle][color]
plt.title('Our First Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20},)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xticks([1,2,3,4])
plt.yticks([0,2,4,6])
plt.legend()# if there are many lines it helps in distinguishing between them.
plt.tight_layout()# used for padding.
#plt.show()
```
