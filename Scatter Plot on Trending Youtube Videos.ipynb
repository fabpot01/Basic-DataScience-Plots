```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
plt.scatter(x, y, s=100, c='red', marker='X')# s is for size c for color.
plt.show()
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
colors=[7,5,9,7,5,7,2,5,3,7,1,2,8,1,9,2,5,6,7,5]# specifying color for each pair(x,y)
sizes=[209,486,381,255,191,315,185,228,174,538,239,394,399,153,273,293,436,501,397,539]#specifying size to each pair

plt.scatter(x,y,s=sizes,c=colors,cmap='Greens',edgecolor='black',linewidth=1,alpha=0.6)# cmap is for colormaps
cbar=plt.colorbar()# add a label for your color map which tells what these colors represents.
cbar.set_label('Satisfaction')
plt.show()
```

```python
data=pd.read_csv('2019-05-31-data.csv')
```

```python
view_count=data['view_count']
likes=data['likes']
ratio=data['ratio']
plt.scatter(view_count,likes,c=ratio,cmap='summer',edgecolor='black',alpha=0.75,linewidth=1)

cbar=plt.colorbar()
cbar.set_label('Likes/Dislike Ratio')
plt.xscale('log')
plt.yscale('log')
plt.title('Trending Youtube Videos')
plt.xlabel('Views Count')
plt.ylabel('Total Likes')
```
