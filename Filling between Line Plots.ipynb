```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('data1.csv')
```

```python
ages=data['Age']
dev_sal=data['All_Devs']
Py_sal=data['Python']
Js_sal=data['JavaScript']
```

```python
plt.plot(ages,dev_sal,color='#444444',linestyle='--',label="All Devs")
plt.plot(ages,Py_sal,label="Python")
plt.legend()
```

```python
plt.plot(ages,dev_sal,color='#444444',linestyle='--',label="All Devs")
plt.plot(ages,Py_sal,label="Python")
plt.fill_between(ages,Py_sal)# y2 is by default 0
plt.legend()
```

```python
plt.plot(ages,dev_sal,color='#444444',linestyle='--',label="All Devs")
plt.plot(ages,Py_sal,label="Python")
overall_median=57287
plt.fill_between(ages,Py_sal,overall_median)#y2 is overall_median.
plt.legend()
```

```python
plt.plot(ages,dev_sal,color='#444444',linestyle='--',label="All Devs")
plt.plot(ages,Py_sal,label="Python")
overall_median=57287
plt.fill_between(ages,Py_sal,overall_median,where=(Py_sal>overall_median),alpha=0.25)#y2 is overall_median,alpha is the transparency.
plt.legend()
```

```python
plt.plot(ages,dev_sal,color='#444444',linestyle='--',label="All Devs")
plt.plot(ages,Py_sal,label="Python")
overall_median=57287
plt.fill_between(ages,Py_sal,overall_median,where=(Py_sal<overall_median),color='red',alpha=0.25)#y2 is overall_median,alpha is the transparency.
plt.legend()
```

```python
plt.plot(ages,dev_sal,color='#444444',linestyle='--',label="All Devs")
plt.plot(ages,Py_sal,label="Python")
overall_median=57287
plt.fill_between(ages,Py_sal,overall_median,where=(Py_sal>overall_median),interpolate=True,alpha=0.25)#y2 is overall_median,alpha is the transparency.
plt.fill_between(ages,Py_sal,overall_median,where=(Py_sal<overall_median),interpolate=True,color='red',alpha=0.25)#y2 is overall_median,alpha is the transparency.
plt.legend()
```

```python
plt.plot(ages,dev_sal,color='#444444',linestyle='--',label="All Devs")
plt.plot(ages,Py_sal,label="Python")
plt.fill_between(ages,Py_sal,dev_sal,where=(Py_sal>dev_sal),interpolate=True,alpha=0.25,label='Python salary above all salary')#y2 is dev_sal,alpha is the transparency.
plt.fill_between(ages,Py_sal,dev_sal,where=(Py_sal<dev_sal),interpolate=True,color='red',alpha=0.6,label='Python salary below all salary')#y2 is dev_sal,alpha is the transparency.
plt.legend()
```
