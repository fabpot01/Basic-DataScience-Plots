```python
import matplotlib.pyplot as plt 
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode=[0 , 0 , 0 , 0.1 , 0]
```

```python
plt.pie(slices,labels=labels,explode=explode,wedgeprops={'edgecolor':'black'},shadow=True,startangle=90,autopct='%1.1f%%')# wedgeprops provides some extrafeatures and autopct to display percentages on pie charts.
```
